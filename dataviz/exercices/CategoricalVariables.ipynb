```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

df = pd.read_csv('../data/IBM_HR_ATTRITION.csv')
df.info()
```

```python
sns.catplot(x='Attrition', y='JobSatisfaction', data=df)
```

```python
sns.boxplot(x='Attrition', y='JobSatisfaction', data=df)
```

```python
df.describe()
```

```python
sns.catplot(x='Attrition', y='TotalWorkingYears', data=df)
```

```python
sns.boxplot(x='Attrition', y='TotalWorkingYears', data=df)
```

```python
df = df[df['TotalWorkingYears'] <= 20]
sns.boxplot(x='Attrition', y='TotalWorkingYears', data=df)
```

```python
df['Attrition'] = df['Attrition'].replace('Yes', 1)
df['Attrition'] = df['Attrition'].replace('No', 0)
sns.catplot(x='Attrition', y='EducationField', data=df, kind='bar')
```
